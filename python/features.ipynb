```python
import cv2
import matplotlib.pyplot as plt
file='../data/face.jpg'
im=cv2.imread(file)[...,::-1]
plt.imshow(im)
```

## 

```python
image=cv2.bilateralFilter(im,0,50,10)
plt.imshow(image)
```

```python
image=cv2.bilateralFilter(im,0,30,10)
plt.imshow(image)
```

## 

```python
plt.imshow(cv2.pyrMeanShiftFiltering(im,10,50))
```

## 

```python
gray=cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
plt.imshow(cv2.equalizeHist(gray))
```
