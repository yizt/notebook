## synthesize

```python
from skimage.feature import hog, hessian_matrix,local_binary_pattern
import numpy as np
import cv2


def normalize_gray(gray_im):
    """
    标准化;将任意值转为0~255
    :param gray_im: numpy array [H,W]
    :return: [H,W] np.uint8 [0~255]
    """
    gray_im[np.isnan(gray_im)]=0
    x = gray_im - np.min(gray_im)
    x /= np.max(x)
    x *= 255
#     im_min,im_max=np.min(gray_im),np.max(gray_im)
#     x = (gray_im-im_min) * 255 /(im_max-im_min)
    #x = gray_im
    return x.astype(np.uint8)


def mix_feature(image):
    """
    将图像的灰度图、hog特征、hessian组合成新的图像
    :param image: [H,W,3]
    :return new_image: [H,W,3]
    """
    new_image=np.zeros_like(image)
    # hog
   # _, hog_feature = hog(image, orientations=9, pixels_per_cell=(8, 8),
   #                      cells_per_block=(3, 3), visualize=True)
   # hog_feature = normalize_gray(hog_feature)

    
     # gray
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # hessian
    hessian_feature = hessian_matrix(gray)
    hessian_feature = normalize_gray(hessian_feature[0])
    #
    lbp=local_binary_pattern(gray,8,4.0,method='var')
    lbp=normalize_gray(lbp)

    #new_image = np.stack([gray, hog_feature, hessian_feature], axis=-1)
    new_image[...,0]=gray
   # new_image[...,1]=hog_feature
   # new_image[...,2]=hessian_feature
    return new_image.astype(np.uint8)
 
```
