```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
im = cv2.imread('/Users/admin/Documents/验证码1.jpg')
#plt.imshow(im)
```

```python
def color_mask(hsv, color):
    h, s, v = hsv[:, :, 0], hsv[:, :, 1], hsv[:, :, 2]
    if color == 'gray':
        return np.logical_and(v <= 220, np.logical_and(s <= 43, v >= 46))
    elif color == 'black':
        return v <= 46
    elif color == 'cyan':
        return np.logical_and(np.logical_and(h >= 78, h <= 99),
                              np.logical_and(s >= 43, v >= 46))
```

```python
hsv = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
plt.imshow(color_mask(hsv,'cyan'))
```
